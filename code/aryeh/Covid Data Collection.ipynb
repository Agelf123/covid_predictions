```python
hospitals = pd.read_csv('../../data/Hospitals.csv')
hospitals.columns = hospitals.columns.str.replace('ZIP','zip')
hospitals.head()
```

```python
med = pd.merge(hospitals,zips,on='zip')
med.columns = med.columns.str.replace('county_name','County')
med.head(
)
```

```python
med['County'] = [i + " " + "County" for i in med['County']]
```

```python
parishes = med.loc[med['STATE'] == 'LA']
parishes['County'] = parishes['County'] + ' Parish'
parishes['STATE'] = 'LA'
(med.loc[med['STATE'] == 'LA']) = parishes
med.head()
```

```python
med_beds = med.groupby('County')['BEDS'].sum()
med_hospitals = (med.groupby('County')['NAME']).value_counts().groupby('County').sum()
len(med_hospitals)
```
