```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('~/downloads/combined.csv')
```

```python
df.drop(columns= ['volcano', 'snowfall'], inplace = True)
```

```python
df.head()
```

```python
df['wildfire'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['tornado'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['landslide'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['hurricane'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['earthquake'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['drought'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['heatwave'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['avalanche'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
df['flood'].replace({
    'None' : 0,
    'Low' : 1,
    'low' : 1,
    'Medium': 2,
    'High' : 3
}, inplace = True)
```

```python
df['hazard score'] = df['wildfire'] + df['tornado'] + df['landslide'] + df['hurricane'] + df['earthquake'] + df['drought'] + df['heatwave'] + df['avalanche'] + df['flood']



```
