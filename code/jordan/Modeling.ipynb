```python
X = lassoed.drop(columns = ['total_covid', 'deaths', 'county_x', 'county_y', 'state', 'id', 'population', 
                           'case_per_pop', 'beds_per_case', 'unnamed: 0_y', 'st_num',
                           'cases', 'countyfips', 'key_0', 'county', 'Unnamed: 0'])
y = lassoed['case_per_pop']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 11, stratify = y)
```

```python
X.head()
```

```python
# Standard Scalar

ss = StandardScaler()     # Instantiate

# Standardize X_train and X_test 

X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
pca = PCA(random_state = random_state) # Instantiate the PCA
pca.fit(X_train)
```

```python
Z_train = pca.transform(X_train)
Z_test = pca.transform(X_test)
```

```python
pd.DataFrame(Z_train).describe()
```

```python
var_exp = pca.explained_variance_ratio_
var_exp
```

```python
log = LogisticRegression(solver = 'liblinear')
log.fit(X_train,y_train)

print('Train: ', round(log.score(X_train, y_train)))
print('Test: ', round(log.score(X_test, y_test)))
```

```python
coefs = log.coef_
coef_df = pd.DataFrame(coefs).T
coef_df['coefficients'] = list(X.columns)
coef_df.rename(columns=({0: 'coefficient',
                        'coefficients': 'feature'}), inplace=True)
coef_df.sort_values(by = 'coefficient')
```
